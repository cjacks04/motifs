```python
def iterate_string(x):
  y=x.split(':')
  count=0
  value=y[0] + ':'
  while count<len(y)-1:
    if y[count].strip()!=y[count+1].strip():
      value=value + y[count+1] + ':'
    count = count+1
  return value[:-1]
```
